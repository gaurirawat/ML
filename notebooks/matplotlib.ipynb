```python
plt.plot([2,4,6,8,10,12], [5,15,10,12,18,8])
plt.xlabel('age', fontsize = 15, color = 'red')
plt.ylabel('height', fontsize = 15, color = 'green')

# used to view only a portion of the actual graph 
plt.xlim(4,10)
```

```python
arr = np.random.randn(20)
# plt.plot(arr, color ='green', linestyle = ':', linewidth = 3, marker = '.', markersize =10)
plt.plot(arr, color ='green', linestyle = '--', marker = 'd', markersize =10)
```

```python
plt.scatter(range(0,20), arr, marker = 's')
```

```python
fig = plt.figure()
ax1 =fig.add_axes([0,0.6,1,0.4])
ax2 =fig.add_axes([0,0,0.8,0.4])
fig.show()
```

```python
fig_a = plt.figure(figsize= (8,8))
ax1_a =fig_a.add_subplot(2,1,1) # means inside a grid of 2*1, the plot shall happen in the first cell 
ax1_a.plot([1,2,3,4], [2,4,6,8])

ax2_a =fig_a.add_subplot(212) # means inside a grid of 2*1, the plot shall happen in the second cell 
ax2_a.plot(range(0,20), arr)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])

#this starts midway of the overall figure and the height and width as 40% of the total figure 
ax2 = fig.add_axes([0.5,0.5,0.4,0.4]) 
plt.show()
```

```python
fig = plt.figure(figsize =(10,10))
a = fig.add_subplot(221) #means we are adding the subplot inside the 1st cell ie top left of the 2*2 grid.
a.plot(range(0,20))

b = fig.add_subplot(222)
b.plot(range(5,15))

c = fig.add_subplot(223)
c.plot(range(0,10))
```
